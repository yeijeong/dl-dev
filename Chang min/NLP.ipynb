```python
import json
import os
from tqdm import tqdm
```

```python
path = './모두의말뭉치/NIKLNEWSPAPER_2022_v1/'
file_path = os.listdir(path)
unit = 'paragraph'
```

```python
files_list = [file for file in file_path]
print(files_list)
```

```python
paragraph_list = []
for file_name in files_list:
    if file_name.endswith(('.json','.JSON')):
        NLP_file = open(path+file_name,'r')
        NLP_data = json.loads(NLP_file.read())
            # 현재 읽고 있는 파일에 포함된 문서를 하나씩 처리하기
        for document in NLP_data['document']:

        # 현재 읽고 있는 문서에 포함된 문단들을 기존 리스트에 덧붙이기
            paragraph_list.extend(document[unit])
```
