---
###### get_market_ticker_list() : 지정한 일자의 코스피 시장에 상장된 ticker를 리스트로 반환

###### 날짜 지정안하면 내부적으로 최근 영업일을 계산해서 ticker 반환
---

```python
# Stock 모듈
tickers = stock.get_market_ticker_list('20200225')
print(tickers)
```
