```python
from tqdm.notebook import tqdm
import scipy
import numpy as np
import pandas as pd

df_feature_market = pd.read_csv(os.path.join('data', 'training', 'qldm_2022a', 'feature', 'feature_market.csv'), index_col=False, converters={'date': lambda x: str(x)})
df_training_market = df_feature_market.fillna(method='ffill')
df_training_market = df_training_market.fillna(0)
file_path = os.path.join('data', 'training', 'qldm_2022a', 'trainingdata', 'trainingdata_market.csv')
os.makedirs(os.path.dirname(file_path), exist_ok=True)
df_training_market.to_csv(file_path, index=False)

for path, dirs, files in os.walk(os.path.join('data', 'training', 'qlds_2022a', 'feature')):
    pbar = tqdm(files)
    for filename in pbar:
        code = filename.split('.')[0].split('_')[-2]
        name = filename.split('.')[0].split('_')[-1]
        pbar.set_description(code)
        file_path = os.path.join(path, filename)
        df_feature = pd.read_csv(file_path, index_col=False, converters={k: lambda x: str(x) for k in ['code', 'name', 'date']})

        # preporcess
        df_training = df_feature.fillna(method='ffill')
        df_training = df_training.fillna(0)

        # find peaks
        peaks = []
        peaks.extend(scipy.signal.find_peaks(df_training['close'], distance=5, width=10)[0])
        peaks.extend(scipy.signal.find_peaks(-df_training['close'], distance=5, width=10)[0])
        if len(df_training)-1 not in peaks:
            peaks.append(len(df_training)-1)

        # 수익성, 안전성, 유동성 지표 추가
        df_training.loc[:, 'peak_date'] = ''
        df_training.loc[:, 'peak_close'] = np.nan
        df_training.loc[:, 'peak_diffratio'] = np.nan
        df_training.loc[:, 'interpeak_mdd'] = np.nan
        df_training.loc[:, 'interpeak_trans_price_exp'] = np.nan

        _last_date = ''  # 이전 날짜를 저장하는 변수 초기화
        for peak in peaks:  # peaks 리스트의 각 peak에 대해 반복
            _date = df_training.iloc[peak]['date']  # 현재 peak의 날짜 저장
            _close = df_training.iloc[peak]['close']  # 현재 peak의 종가 저장
            mask = (df_training['date'] >= _last_date) & (df_training['date'] <= _date)  # 이전 날짜부터 현재 peak 날짜까지의 mask 생성
            _last_date = _date  # 이전 날짜를 현재 peak의 날짜로 업데이트
            if len(df_training[mask]) > 0:  # mask에 해당하는 데이터가 있는 경우
                df_training.loc[mask, 'peak_date'] = _date  # mask에 해당하는 행의 'peak_date' 열에 현재 peak의 날짜 저장
                df_training.loc[mask, 'peak_close'] = _close  # mask에 해당하는 행의 'peak_close' 열에 현재 peak의 종가 저장
                _x = np.array(df_training.loc[mask, 'close'])  # mask에 해당하는 행의 'close' 열을 numpy 배열로 변환하여 _x에 저장
                lower = np.argmax(np.maximum.accumulate(_x) - _x)  # 최대값을 누적한 후 현재 값과의 차이 중 최대값의 인덱스를 lower에 저장
                upper = np.argmax(_x[:lower+1])  # lower 이전까지의 값 중 최대값의 인덱스를 upper에 저장
                df_training.loc[mask, 'interpeak_mdd'] = (_x[lower] - _x[upper]) / _x[upper]  # mask에 해당하는 행의 'interpeak_mdd' 열에 MDD 계산 결과 저장
                df_training.loc[mask, 'interpeak_trans_price_exp'] = df_training.loc[mask, 'trans_price_exp'].mean()  # mask에 해당하는 행의 'interpeak_trans_price_exp' 열에 'trans_price_exp' 열의 평균값 저장
        
        df_training.loc[:, 'peak_diffratio'] = (df_training.loc[:, 'peak_close'] - df_training.loc[:, 'close']) / df_training.loc[:, 'close']  # 'peak_diffratio' 열에 종가와 peak 종가의 차이 비율 계산하여 저장
        df_training = df_training.drop(['peak_date', 'peak_close'], axis=1)  # 'peak_date'와 'peak_close' 열 삭제
        
        file_path = os.path.join('data', 'training', 'qlds_2022a', 'trainingdata', f'trainingdata_{code}_{name}.csv')  # 파일 경로 생성
        os.makedirs(os.path.dirname(file_path), exist_ok=True)  # 파일 경로의 디렉토리 생성
        df_training.to_csv(file_path, index=False)  # DataFrame을 CSV 파일로 저장
```

```python
import pandas as pd


def resample_ticks_to_minutes(ticks, interval):
    # 틱 데이터를 분 단위로 맞추는 함수
    # ticks: 틱 데이터가 담긴 데이터프레임
    # interval: 분 단위로 맞추고자 하는 간격
    
    # 틱 데이터의 시간 정보를 datetime 형식으로 변환
    ticks['time'] = pd.to_datetime(ticks['time'])
    
    # 틱 데이터를 시간별로 그룹화하여 각 그룹의 첫 번째 틱 데이터를 선택
    grouped_ticks = ticks.groupby(pd.Grouper(key='time', freq='H')).first()
    
    # 선택된 틱 데이터의 시간 정보를 분 단위로 맞춤
    resampled_ticks = grouped_ticks.resample(f'{interval}T').ffill()
    
    return resampled_ticks
```

```python
from tqdm.notebook import tqdm
import scipy
import numpy as np
import pandas as pd

df_training_market = pd.read_csv(os.path.join('data', 'training', 'qldm_2022a', 'trainingdata', 'trainingdata_market.csv'), index_col=False, converters={'date': lambda x: str(x)})

X_samples = None
X_label_samples = None
for path, dirs, files in os.walk(os.path.join('data', 'training', 'qlds_2022a', 'trainingdata')):
    pbar = tqdm(files)
    for filename in pbar:
        code = filename.split('.')[0].split('_')[-2]
        name = filename.split('.')[0].split('_')[-1]
        pbar.set_description(code)
        file_path = os.path.join(path, filename)
        df_training = pd.read_csv(file_path, index_col=False, converters={k: lambda x: str(x) for k in ['code', 'name', 'date']})
        df = pd.merge(df_training, df_training_market, on='date', how='left')
        df_label = df[['peak_diffratio', 'interpeak_mdd', 'interpeak_trans_price_exp']].copy()
        df = df.drop(df.columns[:8].tolist() + ['peak_diffratio', 'interpeak_mdd', 'interpeak_trans_price_exp'], axis=1)

        _X = df.values
        _X_label = df_label.values
        _X_samples = _X[np.random.choice(_X.shape[0], 100), :]
        _X_label_samples = _X_label[np.random.choice(_X_label.shape[0], 100), :]
        if X_samples is None:
            X_samples = _X_samples
        else:
            try:
                X_samples = np.concatenate((X_samples, _X_samples))
            except Exception as e:
                print(str(e))
        if X_label_samples is None:
            X_label_samples = _X_label_samples
        else:
            try:
                X_label_samples = np.concatenate((X_label_samples, _X_label_samples))
            except Exception as e:
                print(str(e))

from sklearn.preprocessing import StandardScaler, MaxAbsScaler
scaler_std = StandardScaler()
scaler_std.fit(X_samples)
scaler_label_std = StandardScaler()
scaler_label_std.fit(X_label_samples)
scaler_norm = MaxAbsScaler()
scaler_norm.fit(scaler_std.transform(X_samples))
scaler_label_norm = MaxAbsScaler()
scaler_label_norm.fit(scaler_label_std.transform(X_label_samples))

import pickle
os.makedirs('models', exist_ok=True)
for name, scaler in zip(['scaler_std', 'scaler_label_std', 'scaler_norm', 'scaler_label_norm'], [scaler_std, scaler_label_std, scaler_norm, scaler_label_norm]):
    path = os.path.join('models', f'{name}.pkl')
    with open(path, 'wb') as fout:
        pickle.dump(scaler, fout)
```
