```python
import time
import json
import requests
import pandas as pd

def get_spx_page(start_date, end_date, offset=0):
    # https://api.nasdaq.com/api/quote/SPX/historical?assetclass=index&fromdate=2012-09-23&limit=18&offset=18&todate=2022-09-23
    url = 'https://api.nasdaq.com/api/quote/SPX/historical?assetclass=index&fromdate={start_date}&limit=18&offset={offset}&todate={end_date}'
    headers = {
        'accept': 'application/json, text/plain, */*',
        'accept-encoding': 'gzip, deflate, br',
        'accept-language': 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7',
        'cache-control': 'no-cache',
        'origin': 'https://www.nasdaq.com',
        'pragma': 'no-cache',
        'referer': 'https://www.nasdaq.com/',
        'sec-ch-ua': '"Whale";v="3", " Not;A Brand";v="99", "Chromium";v="104"',
        'sec-ch-ua-mobile': '?0',
        'sec-ch-ua-platform': '"Windows"',
        'sec-fetch-dest': 'empty',
        'sec-fetch-mode': 'cors',
        'sec-fetch-site': 'same-site',
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.122 Whale/3.16.138.27 Safari/537.36',
    }
    res = requests.get(url.format(start_date=f'{start_date[:4]}-{start_date[4:6]}-{start_date[6:]}', end_date=f'{end_date[:4]}-{end_date[4:6]}-{end_date[6:]}', offset=offset), headers=headers)
    data = json.loads(res.text)['data']['tradesTable']['rows']
    df = pd.DataFrame(data)
    if df is None or len(df) == 0:
        return None
    df['date'] = df['date'].apply(lambda x: f'{x[6:]}{x[:2]}{x[3:5]}')
    for k in ['close', 'open', 'high', 'low']:
        df[k] = df[k].astype(float)
    del df['volume']
    return df


def get_spx(start_date, end_date):
    df = pd.DataFrame()
    offset = 0
    last_date = ''
    while last_date == '' or start_date < last_date:
        _df = get_spx_page(start_date, end_date, offset=offset)
        if _df is None or len(_df) == 0 or _df.iloc[-1]['date'] == last_date:
            break
        last_date = _df.iloc[-1]['date']
        df = pd.concat([df, _df], ignore_index=True)
        offset += 18
        time.sleep(1)
    return df


get_spx('20220101', '20220922')
```
