```python
import pandas as pd
import numpy as np
from glob import glob
import os
from datetime import datetime 
```

```python
file_csv = glob("data/A*.csv") # data 폴더 내의 모든 csv파일 목록을 불러오기

total = pd.concat([pd.read_csv(f).assign(filename=os.path.basename(f).split('.csv')[0]) for f in file_csv])
total.sort_values(['filename','date','time'], inplace=True) # 내림차순 정렬
total.rename(columns={'filename':'name'}, inplace=True)
total.reset_index(inplace=True)
total.drop('index', axis=1, inplace=True)


total.to_csv("data/total.csv", encoding='utf-8')
print("1분봉 엑셀 data 취합 완료")
```

```python
df = pd.read_csv('data/total.csv', encoding='utf-8')
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
len(df['name'].unique())
```

```python
df = df.astype({'date':'string', 'time':'string'})

list = []

for ti in df['time']:
    list.append(len(ti))

df['len'] = list

df.loc[ df['len']==3, 'time'] = '0' + df['time']
df.drop('len', axis=1, inplace=True)

df['date_time'] = df['date'] + ' ' + df['time']

df['date_time'] = pd.to_datetime(df['date_time'])

df = df[['date_time', 'name', 'open', 'high', 'low', 'close', 'volume','Accumulated sales volume', 'Accumulated number of contract']]
df.to_csv("data/total.csv", encoding='utf-8')
df
```
