```python
total = pd.read_csv('./A007120.csv')
total.sort_values(['date','time'], inplace=True) # 내림차순 정렬
total.rename(columns={'filename':'name'}, inplace=True)
total.reset_index(inplace=True)
total.drop('index', axis=1, inplace=True)
total
```

```python
import ta
import talib
import numpy as np
import pandas as pd
import scipy
from tqdm.notebook import tqdm
H, L, C, V = total['high'], total['low'], total['close'], total['volume']

total['MFI'] = ta.volume.money_flow_index(
    high=H, low=L, close=C, volume=V, fillna=True)

total['ADI'] = ta.volume.acc_dist_index(
    high=H, low=L, close=C, volume=V, fillna=True)

total['OBV'] = ta.volume.on_balance_volume(close=C, volume=V, fillna=True)
total['CMF'] = ta.volume.chaikin_money_flow(
    high=H, low=L, close=C, volume=V, fillna=True)

total['FI'] = ta.volume.force_index(close=C, volume=V, fillna=True)
total['EOM, EMV'] = ta.volume.ease_of_movement(
    high=H, low=L, volume=V, fillna=True)

total['VPT'] = ta.volume.volume_price_trend(close=C, volume=V, fillna=True)
total['NVI'] = ta.volume.negative_volume_index(close=C, volume=V, fillna=True)
total['VMAP'] = ta.volume.volume_weighted_average_price(
    high=H, low=L, close=C, volume=V, fillna=True)

# Volatility
total['ATR'] = ta.volatility.average_true_range(
    high=H, low=L, close=C, fillna=True)
total['BHB'] = ta.volatility.bollinger_hband(close=C, fillna=True)
total['BLB'] = ta.volatility.bollinger_lband(close=C, fillna=True)
total['KCH'] = ta.volatility.keltner_channel_hband(
    high=H, low=L, close=C, fillna=True)
total['KCL'] = ta.volatility.keltner_channel_lband(
    high=H, low=L, close=C, fillna=True)
total['KCM'] = ta.volatility.keltner_channel_mband(
    high=H, low=L, close=C, fillna=True)
total['DCH'] = ta.volatility.donchian_channel_hband(
    high=H, low=L, close=C, fillna=True)
total['DCL'] = ta.volatility.donchian_channel_lband(
    high=H, low=L, close=C, fillna=True)
total['DCM'] = ta.volatility.donchian_channel_mband(
    high=H, low=L, close=C, fillna=True)
total['UI'] = ta.volatility.ulcer_index(close=C, fillna=True)
# Trend
# total['SMA'] = ta.trend.sma_indicator(close=C, fillna=True)
# total['EMA'] = ta.trend.ema_indicator(close=C, fillna=True)
total['WMA'] = ta.trend.wma_indicator(close=C, fillna=True)
total['MACD'] = ta.trend.macd(close=C, fillna=True)
total['ADX'] = ta.trend.adx(high=H, low=L, close=C, fillna=True)
total['-VI'] = ta.trend.vortex_indicator_neg(
    high=H, low=L, close=C, fillna=True)
total['+VI'] = ta.trend.vortex_indicator_pos(
    high=H, low=L, close=C, fillna=True)
total['TRIX'] = ta.trend.trix(close=C, fillna=True)
total['MI'] = ta.trend.mass_index(high=H, low=L, fillna=True)
total['CCI'] = ta.trend.cci(high=H, low=L, close=C, fillna=True)
total['DPO'] = ta.trend.dpo(close=C, fillna=True)
total['KST'] = ta.trend.kst(close=C, fillna=True)
total['Ichimoku'] = ta.trend.ichimoku_a(high=H, low=L, fillna=True)
total['Parabolic SAR'] = ta.trend.psar_down(
    high=H, low=L, close=C, fillna=True)
total['STC'] = ta.trend.stc(close=C, fillna=True)
# Momentum
total['RSI'] = ta.momentum.rsi(close=C, fillna=True)
total['SRSI'] = ta.momentum.stochrsi(close=C, fillna=True)
total['TSI'] = ta.momentum.tsi(close=C, fillna=True)
total['UO'] = ta.momentum.ultimate_oscillator(
    high=H, low=L, close=C, fillna=True)
total['SR'] = ta.momentum.stoch(close=C, high=H, low=L, fillna=True)
total['WR'] = ta.momentum.williams_r(high=H, low=L, close=C, fillna=True)
total['AO'] = ta.momentum.awesome_oscillator(high=H, low=L, fillna=True)
total['KAMA'] = ta.momentum.kama(close=C, fillna=True)
total['ROC'] = ta.momentum.roc(close=C, fillna=True)
total['PPO'] = ta.momentum.ppo(close=C, fillna=True)
total['PVO'] = ta.momentum.pvo(volume=V, fillna=True)

#등락률 계산
total["close yesterday"] = total["close"].shift(1)
total["change"] = ((total["close"] - total["close yesterday"]) / total["close yesterday"])*100
total["ratio high close"] = ((total["high"] - total["close yesterday"]) / total["close yesterday"])*100
total["ratio low close"] = ((total["low"] - total["close yesterday"]) / total["close yesterday"])*100
total["ratio open close"] = (total["open"] - total["close yesterday"]) / total["close yesterday"]*100
total.drop("close yesterday", axis=1).to_clipboard(index=False)
#이동평균선
total['MA5'] = talib.SMA(total['close'], 5)
total['MA10'] = talib.SMA(total['close'], 10)
total['MA20'] = talib.SMA(total['close'], 20)
total['MA60'] = talib.SMA(total['close'], 60)
total['MA120'] = talib.SMA(total['close'], 120)

#지수평균선
total['EMA5'] = talib.EMA(total['close'], 5)
total['EMA10'] = talib.EMA(total['close'], 12)
total['EMA20'] = talib.EMA(total['close'], 20)
total['EMA60'] = talib.EMA(total['close'], 30)
total['EMA120'] = talib.EMA(total['close'], 90)

total['ratio close MA5'] = total["close"] / total["MA5"]
total['ratio close MA10'] = total["close"] / total["MA10"]
total['ratio close MA20'] = total["close"] / total["MA20"]
total['ratio close MA60'] = total["close"] / total["MA60"]
total['ratio close MA120'] = total["close"] / total["MA120"]

#이동거래평균선
total['MA5 VOLUME'] = talib.SMA(total['volume'], 5)
total['MA10 VOLUME'] = talib.SMA(total['volume'], 10)
total['MA20 VOLUME'] = talib.SMA(total['volume'], 20)
total['MA60 VOLUME'] = talib.SMA(total['volume'], 60)
total['MA120 VOLUME'] = talib.SMA(total['volume'], 120)

# find peaks
peaks = []
peaks.extend(scipy.signal.find_peaks(total['close'], distance=5, width=10)[0])
peaks.extend(scipy.signal.find_peaks(-total['close'], distance=5, width=10)[0])
if len(total)-1 not in peaks:
    peaks.append(len(total)-1)

#수익성, 안정성, 유동성 지표 추가
total['peak_date'] = ''
total['peak_close'] = np.nan
total['peak_diffratio'] = np.nan
total['interpeak_mdd'] = np.nan
total['interpeak_trans_price_exp'] = np.nan

_last_date = ''  # 이전 날짜를 저장하는 변수 초기화
for peak in peaks:  # peaks 리스트의 각 peak에 대해 반복
    _date = total.iloc[peak]['date']  # 현재 peak의 날짜 저장
    _close = total.iloc[peak]['close']  # 현재 peak의 종가 저장
    mask = (total['date'] >= _last_date) & (total['date'] <= _date)  # 이전 날짜부터 현재 peak 날짜까지의 mask 생성
    _last_date = _date  # 이전 날짜를 현재 peak의 날짜로 업데이트
    if len(total[mask]) > 0:  # mask에 해당하는 데이터가 있는 경우
        total.loc[mask, 'peak_date'] = _date  # mask에 해당하는 행의 'peak_date' 열에 현재 peak의 날짜 저장
        total.loc[mask, 'peak_close'] = _close  # mask에 해당하는 행의 'peak_close' 열에 현재 peak의 종가 저장
        _x = np.array(total.loc[mask, 'close'])  # mask에 해당하는 행의 'close' 열을 numpy 배열로 변환하여 _x에 저장
        lower = np.argmax(np.maximum.accumulate(_x) - _x)  # 최대값을 누적한 후 현재 값과의 차이 중 최대값의 인덱스를 lower에 저장
        upper = np.argmax(_x[:lower+1])  # lower 이전까지의 값 중 최대값의 인덱스를 upper에 저장
        total.loc[mask, 'interpeak_mdd'] = (_x[lower] - _x[upper]) / _x[upper]  # mask에 해당하는 행의 'interpeak_mdd' 열에 MDD 계산 결과 저장
        total.loc[mask, 'interpeak_trans_price_exp'] = total.loc[mask, 'trans_price_exp'].mean()  # mask에 해당하는 행의 'interpeak_trans_price_exp' 열에 'trans_price_exp' 열의 평균값 저장
        
total.loc[:, 'peak_diffratio'] = (total.loc[:, 'peak_close'] - total.loc[:, 'close']) / total.loc[:, 'close']  # 'peak_diffratio' 열에 종가와 peak 종가의 차이 비율 계산하여 저장
total = total.drop(['peak_date', 'peak_close'], axis=1)  # 'peak_date'와 'peak_close' 열 삭제



total
```
