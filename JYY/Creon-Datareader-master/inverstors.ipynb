```python
import win32com.client
```

```python
obj = win32com.client.Dispatch('CpSysDib.CpSvr7254')

obj.SetInputValue(0, 'A005930')
obj.SetInputValue(1, 6) #6: 일자별 조회
obj.SetInputValue(4, ord('0')) #0: 내용 전체
obj.SetInputValue(5, 0)
obj.SetInputValue(6, ord('1')) # 1: 순매수량

obj.BlockRequest()
```

```python
cnt = obj.GetHeaderValue(1)
cnt
```

```python
keys = [
    '일자',
    '개인',
    '외국인',
    '기관계',
    '금융투자',
    '보험',
    '투신',
    '은행',
    '기타금융',
    '연기금 등',
    '기타법인',
    '기타외인',
    '사모펀드',
    '국가지자체',
    '종가',
    '대비',
    '대비율',
    '거래량',
    '확정치여부',
]
```

```python
data = []
for i in range(cnt):
    item = {k: obj.GetDataValue(j, i) for j, k in enumerate(keys)}
    data.append(item)
```

```python
data
```

```python
obj.BlockRequest()
for i in range(cnt):
    item = {k: obj.GetDataValue(j, i) for j, k in enumerate(keys)}
    data.append(item)
```

```python
data
```

```python
import pandas as pd
pd.DataFrame(data)
df = pd.DataFrame(data)
df.sort_values(by='일자', ascending=True, inplace=True)
df
```
